```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer,WordNetLemmatizer
from nltk.corpus import stopwords
import re
import pandas as pd
```

```python
data = pd.read_csv("tripadvisor_hotel_reviews.csv")
```

```python
data.info()
```

```python
data.head()
```

```python
data["Review"][0]
```

```python
data["review_lowercase"] = data["Review"].str.lower()
```

```python
data.head()
```

Remove stopwords

```python
en_stopwords = stopwords.words('english')
```

```python
en_stopwords.remove('not')
```

```python
data["review_no_stopwords"]= data["review_lowercase"]\
    .apply(lambda x: " ".join(word for word in x.split()\
                              if word not in en_stopwords))
```

```python
data["review_no_stopwords"][0]
```

```python
data["review_no_stopwords_no_punct"] = data.apply(
    lambda x : re.sub(r"[*]","star",x["review_no_stopwords"]),
    axis=1
)
```

```python
data.head()
```

```python
data["review_no_stopwords_no_punct"]= data.apply(lambda x:re.sub(r"([^\w\s])",'',x["review_no_stopwords_no_punct"]),
                                                 axis=1)
```

```python
data.head()
```

```python
data["tokenized"]= data.apply(lambda x: word_tokenize(x["review_no_stopwords_no_punct"]),
                              axis=1)
```

```python
data.head(3)
```

```python
ps = PorterStemmer()
```

```python
data["stemmed"] = data["tokenized"].apply(lambda tokens :\
                                          [ps.stem(token) for token in tokens])
```

```python
data.head(3)
```

```python
lemetizer= WordNetLemmatizer()
```

```python
data["lemetized"] = data["tokenized"].apply(
    lambda tokens:[lemetizer.lemmatize(token) for token in tokens])
```

```python
data["lemetized"][0]
```

## Let us check N-Grams

```python
tokens_clearn = sum(data["lemetized"],[])
tokens_clearn
```

```python
unigrams = (pd.Series(nltk.ngrams(tokens_clearn,1)).value_counts())
print(unigrams)
```

```python
Bigrams = (pd.Series(nltk.ngrams(tokens_clearn,2)).value_counts())
print(Bigrams)
```
